### a) Dictionary is a Mapping Type

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

merged = dict1.copy()
merged.update(dict2)

print("Merged Dictionary:", merged)
```

```python
keys = ["id", "name", "age"]
values = [101, "Ravi", 21]

student = dict(zip(keys, values))
print("Student Dictionary:", student)
```

```python
data = {"a": 10, "b": 20, "c": 10, "d": 30}

unique_data = {}
for key, value in data.items():
    if value not in unique_data.values():
        unique_data[key] = value

print("After Removing Duplicates:", unique_data)
```

```python
# with dictionary comprehension
original = {
  "United States": "New York", 
  "Italy": "Naples" 
}
inverted = {value: key for key, value in original.items()}
print(inverted)
```

```python
# Without Comprehension
original = {
  "United States": "New York", 
  "Italy": "Naples" 
}
print("Original Dict", original)
inverted_dict = {}
for key, value in original.items():
    inverted_dict[value] = key
    
print("Inverted Dict", inverted_dict)
```

```python
marks = {"Alice": 85, "Bob": 92, "Charlie": 78}

topper = max(marks, key=marks.get)
print("Topper is:", topper)
```

```python
my_dict = {1: "Hello", "Hi": 25, "Howdy": 100}

print(1 in my_dict) 

# the not in operator checks whether key doesn't exist
print("Howdy" not in my_dict) 

print("Hello" in my_dict) 
```

### b) Sets

```python
sentence = "Python is easy and Python is powerful"
words = sentence.split()
unique_words = set(words)
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```

```python
# Create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Finding common elements
common = set1.intersection(set2)

# Finding elements present only in the first set
only_in_first = set1.difference(set2)

# Displaying results
print("Set 1:", set1)
print("Set 2:", set2)
print("Common Elements:", common)
print("Elements only in Set 1:", only_in_first)
```

```python
# Write a program to find common values between two arrays

import numpy as np 
ar1 = [12, 14, 15, 16, 17]
ar2 = [2, 4, 5, 6, 7, 8, 9, 12, 15] 
# Common values between two arrays 
print(np.intersect1d(ar1, ar2))
```
