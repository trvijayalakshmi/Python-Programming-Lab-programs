```python
p = 69
for i in range(5, 0, -1):  
    for j in range(i):   
        print(chr(p), end=" ") 
    print()
    p -= 1
```

```python
n = int(input("Enter n: "))
a, b = 0, 1
print("Fibonacci Series:", end=" ")
for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
```

```python
n = int(input("Enter n: "))
a, b = 0, 1
count = 0
while count < n:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
```

```python
a = 64
b = 4
print(f"The gcd of {a} and {b} is", end=" ")
while b:
    a, b = b, a % b
print(a)
```

```python
total = 0
for i in range(1, 11):
    total += i
print("Sum of first 10 numbers is:", total)
```

```python
s = 6
e = 30
for i in range(s, e):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                if i % j == 0:
                    break
            else:
                print(i,end = " ")
```

```python
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num*i}")
```

```python
num = int(input("Enter a number: "))
fact = 1
i = 1
while i <= num:
    fact *= i
    i += 1
print("Factorial is:", fact)
```

```python
num = int(input("Enter a number: "))
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
print("Reversed number is:", rev)
```

```python
text = input("Enter a string: ")
rev = ""
for ch in text:
    rev = ch + rev
#     rev = rev + ch
print("Reversed String:", rev)
```
