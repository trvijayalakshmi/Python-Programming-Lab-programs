### 1. Type Casting

    • Implicit Type Casting → Done automatically by Python.
    • Explicit Type Casting → Done manually by the programmer using built-in functions.

#### a. Implicit Type Casting

```python
x = 10     
y = 3.5    
z = x + y  # int + float → float
print(z)   
print(type(z)) 
```

#### b. Explicit Type casting

```python
# Converting floating point to integer
x = int(3.14)
print(x) 
print(type(x))

# Converting string to floating point
x = float("3.14")
print(x) 
print(type(x))

# Converting integer to string
x = str(123)
print(x)
print(type(x))
```

### 2. Operators

```python
# 1. Arithmetic Operators

x = 10
y = 5
result = x + y * 2 
print(result)
```

```python
# 2. Relational Operators

a, b = 10, 20
print("a > b :", a > b)     # False
print("a < b :", a < b)     # True

x, y = 15, 15
print("x == y :", x == y)   # True
print("x != y :", x != y)   # False
```

```python
# 3. Logical operators
x, y = 10, 0
print("x and y:", x and y)   # 0 (returns y since x is truthy)
print("x or y:", x or y)     # 10 (returns first truthy value)
```

```python
# 4. Assignment operators

x = 20
x /= 5  
print("Result", x)

x = 17
x %= 5 
print("Result", x)
```

```python
# 5. Membership operators

x = [1, 2, 3, 4, 5]
print(3 in x)  # Output: True
print(6 not in x)  # Output: True
```

```python
# 6. Identity operators

x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)   # True → contents are the same
print(x is y)   # False → stored at different memory locations
print(x is not y)
```

```python
# 7. Bit-wise operators

a = 12  # 1100
b = 6   # 0110

print("a & b =", a & b)   # 0100 → 4        (a and b)
print("a | b =", a | b)   # 1110 → 14     (a or b)
print("a ^ b =", a ^ b)   # 1010 → 10     (a xor b)
print("a << 2 =", a << 2) # 110000 → 48   (Left shift)
print("a >> 2 =", a >> 2) # 0011 → 3    (Right shift)
```

### 3. Input and Output functions

```python
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the annual interest rate (in percentage): "))
time = int(input("Enter the number of periods: "))

amount = principal * (1 + rate/100) ** time
interest = amount - principal

print("The compound interest is:", interest)
print("The compound interest is:", format(interest, ".2f")) 

print(f"The compound interest is: {interest:.2f}")

print(f"{interest} is the compound interest")
print(f"{interest:.2f} is the compound interest")
```
