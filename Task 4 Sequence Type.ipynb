### Seqeunce Type - Stings, Lists and Tuples

### a) Strings

```python
word = "Apple"
result = ""
for i, char in enumerate(word):
    result += char
    if i < len(word) - 1:
        result += ","
print(result)
```

```python
word = "Apple"
result = ",".join(word)
print(result)
```

```python
s = "This is a sample string with sample words."
rem = "sample"
result = s.replace(rem, "")  
print(result)

new_text = s.replace(" ", "-") 
print(new_text)   # replaces all spaces with -
```

### b) Lists

```python
s = input("Enter a sentence: ")
words = s.split()
reversed_words = [word[::-1] for word in words]
print(" ".join(reversed_words))
```

```python
squares = [x**2 for x in range(1, 6)]
print(squares)  
```

```python
nums = [x for x in range(1, 21) if x % 2 == 0 and x % 3 == 0]
print(nums)
```

```python
table = [[i*j for j in range(1, 6)] for i in range(1, 4)]
print("Multiplication table:", table)
```

### c) Tuples

```python
x = ("apple", "banana", "cherry")
y = list(x)
y.append("orange")
x = tuple(y)
print(x)
```

```python
x = ("apple", "banana", "cherry")
print(x[1])
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)
```
